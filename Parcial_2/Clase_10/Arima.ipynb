## Que es el modelo ARIMA

- Partimos del punto de pregunta, que es el modelo estacionario de tendencia:
    - Tiene tanto la media la varianza y la covarianza constante.

- Modelo estacionario en **Tendencia**
$$
Z_t = \alpha + \beta t + ARMA(p,q)
$$

- Modelo Estacionario en **Diferencia**

$$
\Delta Z_t = \delta + ARMA(p,q)
$$

- ARMA (P,Q) withs lags

$$
Z_t = \theta_1 Z_t L + \theta_2 Z_t L^2 + \theta_3 Z_t L^3 + .... \theta_p Z_t L^p + \epsilon_t + \pi_1 \epsilon_t L + \pi_2 \epsilon_t L^2 + ... \pi_q \epsilon_t L^q
$$

$$
Z_t =  \frac{\epsilon_t(1 + \pi_1  L + \pi_2  L^2 + ... \pi_q  L^q)}{1 - (\theta_1 L + \theta_2 L^2 + \theta_3 L^3 + .... \theta_p L^p)}
$$

$$
Z_t =  \frac{\epsilon_t(1 + \pi_1  L + \pi_2  L^2 + ... \pi_q  L^q)}{(1-\lambda_1 L)(2-\lambda_2 L)(3-\lambda_3 L)...(1-\lambda_p L)}
$$

- Caso 1

    - $Z_t =  \frac{\epsilon_t(1 + \pi_1  L + \pi_2  L^2 + ... \pi_q  L^q)}{(0)(2-\lambda_2 L)(3-\lambda_3 L)...(1-\lambda_p L)}$

    - $\lambda_1 = 1$

    -  Indeterminado

- Caso 2

    - $Z_t =  \frac{\epsilon_t(1 + \pi_1  L + \pi_2  L^2 + ... \pi_q  L^q)}{(0)(0)(3-\lambda_3 L)...(1-\lambda_p L)}$

    - $\lambda_1 = \lambda_2 = 1$

    -  Indeterminado

- Caso 3

    - $Z_t =  \frac{\epsilon_t(1 + \pi_1  L + \pi_2  L^2 + ... \pi_q  L^q)}{(0)(0)(0)...(1-\lambda_p L)}$

    - $\lambda_1 = \lambda_2 = \lambda_3 = 1$

    -  Indeterminado

- Solucion del Problema

$$
Z_t(1-L)^3 =  \frac{\epsilon_t(1 + \pi_1  L + \pi_2  L^2 + ... \pi_q  L^q)}{(2-\lambda_2 L)(3-\lambda_3 L)...(1-\lambda_p L)}
$$

$$
Z_t(1-L) =  \Psi_1 (L) \epsilon_t
$$

$$
\Delta Z_t =  \Psi_1 (L) \epsilon_t
$$

- Caso 3 ARMA integrada de orden 3

$$
Z_t(1-\lambda_1 L) =  \frac{\epsilon_t(1 + \pi_1  L + \pi_2  L^2 + ... \pi_q  L^q)}{(2-\lambda_2 L)(3-\lambda_3 L)...(1-\lambda_p L)}
$$

$$
Z_t(1 - 3L + 3L^2 - L^3) =  \Psi_3 (L) \epsilon_t
$$

$$
Z_t - 3Z_{t-1} + 3Z_{t-2} - Z_{t-3} =  \Psi_3 (L) \epsilon_t
$$
